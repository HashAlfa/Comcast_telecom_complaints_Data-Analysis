```python
# Data Preprocessing :
```

```python
df.info()
```

```python
# convert datatype of  'Date_month_year' to 'datetime'

df['Date_month_year'] =  df['Date_month_year'].apply(pd.to_datetime)
```

```python
df.info()
```

```python
# convert 'Zip code' into object

df['Zip code'] = df['Zip code'].astype(object)

df.info()
```

```python
df['Status'].unique()
```

```python
# Data Bucketing : 

# List Comprehension

# list_name = [var        if condition else        for i in sequence_name ]

df['new_Status'] = ['Closed'   if i == 'Closed' or i=='Solved'  else  'Open'       for i in df['Status'] ]
```

```python
df.head(10)
```

```python
df['new_Status'].unique()
```

```python
# Data Analysis using data Visualization : 

# 1. Univariate Analysis : 
```

```python
# 1. How many daily complaints received ?

df['Date_month_year'].value_counts()[: 5]
```

```python
df['Date_month_year'].value_counts()[: 10].plot()

plt.title('Daily Compliant Chart')
plt.xlabel('Dates')
plt.ylabel(' No of Complaints Received ')
plt.show()
```

```python
# Give top 10 states with complaints?

df['State'].value_counts()[: 10]
```

```python

df['State'].value_counts()[: 10].plot.bar(figsize = (10,7))
```

```python
# 2. Bivariate analysis : 

# 1. Give statewise complaint status.

# groupby()

df.groupby(['State' ,'new_Status']).size()
```

```python
# to convert into tabular format

df.groupby(['State' ,'new_Status']).size().unstack()
```

```python
# top 5 complaint receiving states.

df.groupby(['State']).size().sort_values(ascending = False)[:5]
```

```python
# Plot Statewise Complaints

statewise_complaint = df.groupby(['State' ,'new_Status']).size().unstack()[:10]

statewise_complaint.plot.bar(figsize = (15,7), stacked = True)
```
